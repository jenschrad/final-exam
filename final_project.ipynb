```python
# this program takes employee names and their salaries
# as input and stores them in parallel lists.
# input ends when the sentinel value is reached.
# once the lists are completed, both lists are displayed.
# then, the average of all salaries is calculated and displayed.
# the program will find the employees who's salaries
# fall within 5,000 of the average and display their names.

# intializes employee name variable
# holds user input
employee_name = ""

# intializes employee salary variable 
# holds user input
employee_salary = 0

# intializes empty list
# for employee names
employees = []

# intializes empty list
# for employee salaries
salaries = []

# sets sentinel value 
sentinel = "End"

# intializes variable that
# holds sum of employee salaries
total_salaries = 0

# intializes variable that
# holds average of employee salaries
average_salary = 0

# while loop for taking user input
# and filling lists
# used while true to circumvent
# needing mulitple input statements
while True:
    employee_name = str(input("Please enter the employee name, when finished enter " + "'" + sentinel+ "'" + ": "))
    # if user enters sentinel, loop breaks
    if employee_name == sentinel:
        break
    # fills employee names list with user input
    employees.append(employee_name)
    
    employee_salary = float(input("Please enter " + employee_name + "'s salary:"))
    
    # fills employee salary list with user input
    salaries.append(employee_salary)

print("\n")

# displays title underlined and bold
print("\033[1;4m" + "Employee Salary Report" + "\033[0m")

print("\n")

# displays employee and salary
# headings left adjusted and bold
print("\033[1m" + "Employee".ljust(15), "Salary" + "\033[0m")

# for loop to display lists 
# lists are displayed left adjusted
for i in range(len(employees)):
    print(employees[i].ljust(15) + "\t" + "$" + str(salaries[i]))
    
# calculates sum of salaries in 
# in employee salary list
for i in range(len(salaries)):
    total_salaries = sum(salaries)
    
# calculates average of salaries
# in employee salary list
average_salary = total_salaries / len(salaries)


print("\n")

# displays heading for 
# average salary in bold
print("\033[1m" + "Average Employee Salary" + "\033[0m")
print("\t$" + str(round(average_salary, 2)))

# variable for low end of average range
min_range = average_salary - 5.0

# variable for high end of average range
max_range = average_salary + 5.0
print("\n")

# displays heading for 
# average salary range in bold
print("\033[1m" +" Average Salary Range" + "\033[0m")
print( "   $" + str(round(min_range, 2)) + " - "  + "$" + str(round(max_range, 2)))


            
print("\n")
# displays heading for employees
# within average range in bold
print("\033[1m" + "  Employees within" + "\n" + "    Salary Range" + "\033[0m")

# loop that iterates through
# salary list to compare
# to average salary range (within 5,000)
# and then displays employee index
# if they are in range
for i in range(len(salaries)):
    if salaries[i] >= min_range and salaries[i] <= max_range:
        print(employees[i].center(20))

    
        

    
    
    
```
